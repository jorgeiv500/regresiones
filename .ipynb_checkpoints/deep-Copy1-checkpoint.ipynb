## üî∏ Introducci√≥n: ¬øPor qu√© necesitamos Deep Learning?

A lo largo de la historia del aprendizaje autom√°tico (*Machine Learning*), hemos desarrollado modelos cl√°sicos como:

* **Regresi√≥n lineal/log√≠stica**
* **√Årboles de decisi√≥n**
* **K-vecinos m√°s cercanos (KNN)**
* **Support Vector Machines (SVM)**

Estos modelos se basan en principios estad√≠sticos y funcionan bastante bien cuando:

* La cantidad de datos es **moderada**.
* Las relaciones entre variables son **lineales o f√°cilmente separables**.
* El **dise√±o de caracter√≠sticas** (feature engineering) es guiado por expertos.

Sin embargo, en la √∫ltima d√©cada se hizo evidente que estos modelos enfrentan **limitaciones** ante:

* Datos no estructurados como im√°genes, audio o texto.
* Vol√∫menes masivos de informaci√≥n.
* Relaciones **altamente no lineales y jer√°rquicas**.

Aqu√≠ es donde entra el **Deep Learning**.

---


---

## üî∏ 3. ¬øQu√© es una Neurona Artificial?

Inspirada en las neuronas biol√≥gicas, una **neurona artificial** es una unidad que:

1. **Recibe entradas** (features) como $x_1, x_2, \ldots, x_n$
2. **Multiplica cada entrada por un peso** $w_i$
3. **Suma los productos y agrega un sesgo (bias)** $b$
4. **Pasa el resultado por una funci√≥n de activaci√≥n** $\phi$

**F√≥rmula general de una neurona:**

$$
y = \phi\left(\sum_{i=1}^{n} w_i x_i + b\right)
$$

Donde $\phi$ puede ser:

* **Lineal** (como en regresi√≥n)
* **Escal√≥n (Heaviside)**: para clasificaci√≥n binaria (Perceptr√≥n)
* **ReLU**: para redes profundas
* **Sigmoide o Tanh**: para salidas entre -1 y 1 o 0 y 1

---



## üî∏ 4. ¬øQu√© es una Red Neuronal?

Una **red neuronal** es un conjunto de neuronas conectadas en capas:

* **Capa de entrada:** donde ingresan los datos.
* **Capas ocultas:** donde ocurren las transformaciones complejas.
* **Capa de salida:** donde se produce la predicci√≥n final.

### üî∏ Perceptr√≥n

El **Perceptr√≥n** es la red neuronal m√°s simple, sin capas ocultas. Fue propuesto por **Frank Rosenblatt** en 1958.

### üî∏ Multilayer Perceptron (MLP)

El **Perceptr√≥n Multicapa (MLP)** es una extensi√≥n que permite tener **una o m√°s capas ocultas**. Esto lo convierte en un modelo mucho m√°s poderoso.

---



## üî∏ 5. ¬øPor qu√© el MLP supera a los modelos cl√°sicos?

Seg√∫n Goodfellow, el MLP:

* Aprende representaciones intermedias (caracter√≠sticas jer√°rquicas).
* Supera a modelos con ingenier√≠a de caracter√≠sticas manual.
* Se entrena eficientemente usando el algoritmo de retropropagaci√≥n (*backpropagation*).

---



## üî∏ 6. Funciones de Activaci√≥n Comunes

| Funci√≥n | F√≥rmula                             | Uso com√∫n                |
| ------- | ----------------------------------- | ------------------------ |
| Sigmoid | $\frac{1}{1 + e^{-x}}$              | Clasificaci√≥n binaria    |
| Tanh    | $\frac{e^x - e^{-x}}{e^x + e^{-x}}$ | Salidas entre -1 y 1     |
| ReLU    | $\max(0, x)$                        | Capas ocultas profundas  |
| Softmax | $\frac{e^{x_i}}{\sum e^{x_j}}$      | Clasificaci√≥n multiclase |

````python
import pandas as pd
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer/breast-cancer.data"
columnas = ["Class", "age", "menopause", "tumor-size", "inv-nodes", "node-caps","deg-malig", "breast", "breast-quad", "irradiat"
]
df = pd.read_csv(url, names=columnas)
df = df.replace('?', pd.NA).dropna()

# Mapear valores ordinales
mapeos = {
    'age': {'10-19': 15, '20-29': 25, '30-39': 35, '40-49': 45,
            '50-59': 55, '60-69': 65, '70-79': 75, '80-89': 85, '90-99': 95},
    'tumor-size': {'0-4': 2, '5-9': 7, '10-14': 12, '15-19': 17, '20-24': 22,
                   '25-29': 27, '30-34': 32, '35-39': 37, '40-44': 42,
                   '45-49': 47, '50-54': 52, '55-59': 57},
    'inv-nodes': {'0-2': 1, '3-5': 4, '6-8': 7, '9-11': 10, '12-14': 13,
                  '15-17': 16, '18-20': 19, '21-23': 22, '24-26': 25,
                  '27-29': 28, '30-32': 31, '33-35': 34, '36-39': 38}
}

for col, mapping in mapeos.items():
    df[col] = df[col].map(mapping)

# Mapear variables binarias
df['Class'] = df['Class'].map({'no-recurrence-events': 0, 'recurrence-events': 1})
df['node-caps'] = df['node-caps'].map({'no': 0, 'yes': 1})
df['irradiat'] = df['irradiat'].map({'no': 0, 'yes': 1})
df['breast'] = df['breast'].map({'left': 0, 'right': 1})
df['deg-malig'] = pd.to_numeric(df['deg-malig'], errors='coerce')
df['tumor-size'] = pd.to_numeric(df['tumor-size'], errors='coerce')
df['inv-nodes'] = pd.to_numeric(df['inv-nodes'], errors='coerce')

df.head()
````

````python
from sklearn.preprocessing import OneHotEncoder
categorical_columns = ['menopause', 'breast', 'breast-quad']
encoder = OneHotEncoder(sparse=False, drop='first')
encoded_data = encoder.fit_transform(df[categorical_columns])
encoded_df = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(categorical_columns))
df_encoded = pd.concat([df.drop(columns=categorical_columns), encoded_df], axis=1)
df_encoded.head()
````

````python
# Verificar si hay valores nulos en el conjunto de datos
df_encoded.isnull().sum()
````

````python
# Imputar los valores nulos en df_encoded

# Para variables num√©ricas, se utiliza la mediana
num_cols = ['age', 'tumor-size', 'inv-nodes', 'deg-malig']
df_encoded[num_cols] = df_encoded[num_cols].fillna(df_encoded[num_cols].median())

# Para variables categ√≥ricas (One Hot Encoded), se utiliza el valor 0 (ausencia de categor√≠a)
cat_cols = [
    'node-caps', 'irradiat', 'menopause_lt40', 'menopause_premeno',
    'breast_1', 'breast-quad_left_low', 'breast-quad_left_up',
    'breast-quad_right_low', 'breast-quad_right_up'
]
df_encoded[cat_cols] = df_encoded[cat_cols].fillna(0)

# Para la columna 'Class', se eliminan las filas con nulos ya que es el objetivo
df_encoded = df_encoded.dropna(subset=['Class'])

# Verificaci√≥n de nulos
df_encoded.isnull().sum()
````

````python
df_encoded.head()
````

````python
# üìö Librer√≠as necesarias
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, roc_auc_score
import matplotlib.pyplot as plt
import pandas as pd

X = df_encoded.drop(columns=['Class'])
y = df_encoded['Class']

# üì¶ Dividir el conjunto de datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# ‚öôÔ∏è Crear y entrenar el modelo de regresi√≥n log√≠stica
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

# üöÄ Predecir en el conjunto de prueba
y_pred = model.predict(X_test)
y_prob = model.predict_proba(X_test)[:, 1]

# üîç Reporte de clasificaci√≥n
print("üîç Reporte de Clasificaci√≥n:")
print(classification_report(y_test, y_pred))

# üìä Matriz de Confusi√≥n
print("üìä Matriz de Confusi√≥n:")
print(confusion_matrix(y_test, y_pred))

# üìà Curva ROC
fpr, tpr, _ = roc_curve(y_test, y_prob)
plt.figure(figsize=(6, 4))
plt.plot(fpr, tpr, label=f"AUC = {roc_auc_score(y_test, y_prob):.2f}")
plt.plot([0, 1], [0, 1], '--', color='gray')
plt.xlabel("Tasa de Falsos Positivos (FPR)")
plt.ylabel("Tasa de Verdaderos Positivos (TPR)")
plt.title("Curva ROC - Regresi√≥n Log√≠stica")
plt.legend()
plt.grid()
plt.show()
````

````python
# Reimportar librer√≠as necesarias
import tensorflow as tf
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, roc_auc_score
import matplotlib.pyplot as plt

# Definir el modelo en TensorFlow
model = tf.keras.Sequential([
    tf.keras.layers.InputLayer(input_shape=(X_train.shape[1],)),
    tf.keras.layers.Dense(10, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

# Compilar el modelo
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Entrenar el modelo
history = model.fit(X_train, y_train, epochs=100, batch_size=16, verbose=0)

# Predecir
y_pred = (model.predict(X_test) > 0.5).astype("int32")
y_prob = model.predict(X_test).flatten()

# Reporte de clasificaci√≥n
print("üîç Reporte de Clasificaci√≥n - MLP (TensorFlow):")
print(classification_report(y_test, y_pred))

# Matriz de confusi√≥n
print("üìä Matriz de Confusi√≥n:")
print(confusion_matrix(y_test, y_pred))

# Curva ROC
fpr, tpr, _ = roc_curve(y_test, y_prob)
plt.figure(figsize=(6, 4))
plt.plot(fpr, tpr, label=f"AUC = {roc_auc_score(y_test, y_prob):.2f}")
plt.plot([0, 1], [0, 1], '--', color='gray')
plt.xlabel("Tasa de Falsos Positivos (FPR)")
plt.ylabel("Tasa de Verdaderos Positivos (TPR)")
plt.title("Curva ROC - MLP (TensorFlow)")
plt.legend()
plt.grid()
plt.show()
````

````python
# Re-importar librer√≠as necesarias
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
import pandas as pd


X = df_encoded.drop(columns=['Class'])
y = df_encoded['Class']

# Crear y entrenar el modelo de regresi√≥n log√≠stica
model = LogisticRegression(max_iter=1000)

# Validaci√≥n cruzada (K-Fold con 5 particiones)
cv_scores = cross_val_score(model, X, y, cv=6, scoring='roc_auc')

# Resultados de la validaci√≥n cruzada
cv_results = pd.DataFrame({
    'Fold': [1, 2, 3, 4, 5,6],
    'ROC AUC': cv_scores
})

# M√©tricas adicionales
mean_auc = cv_scores.mean()
std_auc = cv_scores.std()

# Mostrar resultados
print(cv_results)
print("\nPromedio AUC: ", mean_auc)
print("Desviaci√≥n Est√°ndar AUC: ", std_auc)

````

````python
from sklearn.model_selection import GridSearchCV

# Definir el modelo
model = LogisticRegression(max_iter=1000)

# Definir los hiperpar√°metros a evaluar
param_grid = {
    'C': [0.01, 0.1, 1, 10, 100],
    'solver': ['liblinear', 'saga'],
    'penalty': ['l1', 'l2']
}

# Configuraci√≥n del GridSearchCV
grid_search = GridSearchCV(estimator=model, param_grid=param_grid, 
                           scoring='roc_auc', cv=5, verbose=1, n_jobs=-1)

# Ejecutar b√∫squeda
grid_search.fit(X, y)

# Resultados
best_model = grid_search.best_estimator_
best_params = grid_search.best_params_
best_score = grid_search.best_score_

print("üîç Mejores Hiperpar√°metros: ", best_params)
print("üåü Mejor AUC obtenido: ", best_score)
````

## üìå **Explicaci√≥n del Modelo:**

La **Regresi√≥n Log√≠stica** es un modelo lineal que se utiliza para problemas de clasificaci√≥n binaria. En este caso, buscamos predecir si un paciente tendr√°:

* `0`: **No recurrencia** del c√°ncer.
* `1`: **Recurrencia** del c√°ncer.

La ecuaci√≥n principal es:

$$
P(Y=1|X) = \frac{1}{1 + e^{-(wX + b)}}
$$

Donde:

* $w$ son los pesos asociados a cada caracter√≠stica.
* $X$ son las caracter√≠sticas del paciente (edad, tama√±o del tumor, etc.).
* $b$ es el sesgo o bias.

---

## üîé **¬øQu√© es un Perceptr√≥n?**

Un **Perceptr√≥n** es la unidad m√°s b√°sica de una red neuronal, introducida por **Frank Rosenblatt** en 1958. Se puede considerar como una neurona artificial que:

1. **Recibe entradas** (features) ponderadas.
2. **Suma las entradas** con un sesgo (bias).
3. **Aplica una funci√≥n de activaci√≥n** para producir una salida binaria.

---

### üìå **F√≥rmula Matem√°tica**

La salida del perceptr√≥n se define como:

$$
y = \phi(\mathbf{w} \cdot \mathbf{x} + b)
$$

Donde:

* $\mathbf{w}$ son los pesos asociados a cada entrada.
* $\mathbf{x}$ son los valores de las caracter√≠sticas.
* $b$ es el sesgo (bias).
* $\phi$ es una funci√≥n de activaci√≥n, t√≠picamente un escal√≥n:

$$
\phi(z) = 
\begin{cases} 
1 & \text{si } z \geq 0 \\
0 & \text{si } z < 0
\end{cases}
$$

---

### üîπ **Intuici√≥n Visual**

El perceptr√≥n intenta encontrar un hiperplano que separe las clases. Si las clases son linealmente separables, el perceptr√≥n puede encontrar esa frontera de decisi√≥n sin errores.

# Introducci√≥n a Redes Feedforward Profundas

* Las redes feedforward profundas, tambi√©n conocidas como **Perceptrones Multicapa (MLPs)**, son el modelo b√°sico para el aprendizaje profundo.
* Su objetivo principal es aprender a predecir un valor de salida a partir de una serie de datos de entrada.
* La informaci√≥n siempre se mueve en una sola direcci√≥n (hacia adelante), sin bucles ni retroalimentaci√≥n.

---

# Componentes Clave de la Arquitectura

* Las redes est√°n formadas por tres tipos de capas:

  * **Capa de Entrada**: Donde ingresan los datos. Ejemplo: una imagen, un n√∫mero o un conjunto de caracter√≠sticas.
  * **Capas Ocultas**: Donde se procesa la informaci√≥n, aplicando transformaciones matem√°ticas.
  * **Capa de Salida**: Genera el resultado final, como una predicci√≥n o una clasificaci√≥n.

### Matem√°ticamente:

$$
    a = Wx + b \quad \text{y} \quad h = f(a)
$$

Donde:

* \$W\$ son los pesos (importancia de cada entrada).
* \$x\$ es el dato de entrada.
* \$b\$ es el sesgo (bias) para ajustar el resultado.
* \$f\$ es la funci√≥n de activaci√≥n, que introduce no linealidad.

---

# Ejemplo Pr√°ctico: Clasificaci√≥n con TensorFlow

```{python}
#| code-line-numbers: "|4|8|13"
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense

# Definimos un modelo simple con tres capas
model = Sequential([
    Dense(16, activation='relu', input_shape=(30,)),
    Dense(8, activation='relu'),
    Dense(1, activation='sigmoid')
])

# Compilamos el modelo
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Resumen de la arquitectura
model.summary()
```

* En este ejemplo se crea un modelo con:

  * Una capa de entrada con 30 caracter√≠sticas.
  * Dos capas ocultas (16 y 8 neuronas).
  * Una capa de salida con activaci√≥n `sigmoid` para clasificaci√≥n binaria.

---

````python
df= housing
# An√°lisis inicial: convertir el problema en clasificaci√≥n
# Definimos el umbral como el percentil 75 del valor de la casa para clasificar "caro" vs "barato"
umbral_precio = df['median_house_value'].quantile(0.75)
df['high_price'] = (df['median_house_value'] >= umbral_precio).astype(int)

# Preparar datos para modelos
features = ['housing_median_age', 'total_rooms', 'total_bedrooms', 'population',
            'households', 'median_income', 'longitude', 'latitude']
X = df[features].fillna(df[features].mean())
y = df['high_price']

# Divisi√≥n de entrenamiento y prueba
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Escalado de caracter√≠sticas
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Entrenamiento de modelos: Regresi√≥n Log√≠stica, Perceptr√≥n y MLP
from sklearn.linear_model import LogisticRegression, Perceptron
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report

# Modelos
lr = LogisticRegression(max_iter=1000).fit(X_train_scaled, y_train)
pc = Perceptron(max_iter=1000, tol=1e-3).fit(X_train_scaled, y_train)
mlp = MLPClassifier(hidden_layer_sizes=(10, 5), max_iter=1000, random_state=42).fit(X_train_scaled, y_train)

# Predicciones
pred_lr = lr.predict(X_test_scaled)
pred_pc = pc.predict(X_test_scaled)
pred_mlp = mlp.predict(X_test_scaled)

# Reportes
report_lr = classification_report(y_test, pred_lr, output_dict=True)
report_pc = classification_report(y_test, pred_pc, output_dict=True)
report_mlp = classification_report(y_test, pred_mlp, output_dict=True)

# Convertir a DataFrames para comparar
df_lr = pd.DataFrame(report_lr).transpose().add_prefix('LR_')
df_pc = pd.DataFrame(report_pc).transpose().add_prefix('PC_')
df_mlp = pd.DataFrame(report_mlp).transpose().add_prefix('MLP_')

# Unir resultados clave (accuracy y f1-score)
comparacion = pd.concat([
    df_lr.loc[['accuracy', 'weighted avg']].rename(index={'weighted avg': 'F1'}).iloc[:, [0]],
    df_pc.loc[['accuracy', 'weighted avg']].rename(index={'weighted avg': 'F1'}).iloc[:, [0]],
    df_mlp.loc[['accuracy', 'weighted avg']].rename(index={'weighted avg': 'F1'}).iloc[:, [0]],
], axis=1)

comparacion.columns = ['Regresi√≥n Log√≠stica', 'Perceptr√≥n', 'MLP']

import ace_tools as tools; tools.display_dataframe_to_user(name="Comparaci√≥n de Modelos (Clasificaci√≥n de Precio Alto)", dataframe=comparacion)
````
